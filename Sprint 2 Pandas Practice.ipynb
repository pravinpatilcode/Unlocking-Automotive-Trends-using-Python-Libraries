```python
# TASK 6
import pandas as pd

#  data
maths_marks = pd.Series([35, 86, 63, 90, 50, 63, 15, 72, 15, 54, 72, 83, 58, 70, 92, 40, 59, 63, 53, 59, 55, 85, 18, 83, 67, 85, 46, 59, 77, 71, 77, 56, 79, 61, 82, 76, 60, 52, 88, 78])
physics_marks = pd.Series([85, 94, 63, 59, 76, 64, 87, 55, 51, 71, 54, 94, 33, 72, 75, 67, 85, 15, 50, 50, 19, 55, 59, 65, 76, 13, 57, 70, 51, 94, 55, 50, 71, 91, 75, 51, 90, 70, 87, 56])
student_ID = pd.Series(range(1001, 1041))

# DataFrame from the provided Series
data = pd.DataFrame({
    'student_ID': student_ID,
    'maths_marks': maths_marks,
    'physics_marks': physics_marks
})

# Step 1
#  Filter students who meet the qualification criteria
qualified_students = data[(data['maths_marks'] >= 50) &
                          (data['physics_marks'] >= 50) &
                          (data['maths_marks'] + data['physics_marks'] >= 120)]

# Step 2
#  Create a DataFrame of the qualifying students' IDs
qualifying_df = pd.DataFrame({
    'round 2 qualifiers': qualified_students['student_ID']
})

print(qualifying_df)
```
