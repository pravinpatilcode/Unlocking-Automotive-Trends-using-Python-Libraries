```python
#importing numpy & pandas
import numpy as np
import pandas as pd
```

```python
#  csv file reading

df = pd.read_csv('DS1_C4_S3_SalesData_Practice.csv')
df
```

```python
# TASK 1

df_sorted = df.sort_values(by='Sales', ascending=True, inplace=True)
df
```

```python
# TASK 2
# Displaying the all rows where the sale amount is >= 45000

df_filter = df[df["Sales"] >= 45000]
df_filter
```

```python
# TASK 3

df_California = pd.DataFrame(df[df["State"] == "California"], df.sort_values(by='Sales' ,ascending=True ,inplace=True))
df_California
```

```python
#creating the dataframe for state Texas and sorting the sales values in asc

df_Texas = pd.DataFrame(df[df["State"] == "Texas"], df.sort_values(by='Sales', ascending=True,inplace=True))
df_Texas
```

```python
#creating the dataframe for state Montana and sorting the sales values in asc

df_Montana = pd.DataFrame(df[df["State"] == "Montana"], df.sort_values(by='Sales', ascending=True,inplace=True))
montana_count = len(df_Montana)
print(f"Number of records for Montana: {montana_count}")
df_Montana
```

```python
# TASK 4
state = df_California['State'].sum()
sales_sum_california = df_California['Sales'].sum()
sales_sum_Texas = df_Texas['Sales'].sum()
sales_sum_Montana = df_Montana['Sales'].sum()
print("Total Sales in the state of California:" ,sales_sum_california)
print("Total Sales in the state of California:" ,sales_sum_california)
print("Total Sales in the state of Texas:" ,sales_sum_Texas)
print("Total Sales in the state of Montana:" ,sales_sum_Montana)
```

```python
# TASK 5

max_sales_month = df_California.loc[df_California['Sales'].idxmax(), 'Month']    
#  Find the months with highest and lowest sales
min_sales_month = df_California.loc[df_California['Sales'].idxmin(), 'Month']

#printing the Months with max and min sales 
print(f"Month with the highest sales in California: {max_sales_month}")  # Display the results
print(f"Month with the lowest sales in California: {min_sales_month}")
```

```python
# TASK 6
#Creating a dataframe which contain  States and Sale in thousands information

df1 = pd.DataFrame({   "State":['Alaska','Texas','California','Montana','New Mexico'],
                       "Sale in thousands": [df[df['State'] == 'Alaska']['Sales'].sum(),
                                         df_Texas['Sales'].sum(),
                                        df_California['Sales'].sum(),
                                        df_Montana['Sales'].sum(),
                                       df[df['State'] == 'New Mexico']['Sales'].sum()]  })

df1
```

```python
# TASK 7
df1.sort_values(by = 'Sale in thousands', ascending=True, inplace= True)
df1
```

```python
#saving the df1 dataframe to the local System as total_sales.csv
df1.to_csv('total_sales.csv') #  file is saved to the current working dictionary
```
