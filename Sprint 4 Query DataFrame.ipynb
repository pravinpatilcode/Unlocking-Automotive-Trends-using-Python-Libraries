```python
#importing numpy & pandas
import numpy as np
import pandas as pd
```

```python
# TASK 1
df = pd.read_csv('DS1_C4_S4_Test_Scores_Data_Practice.csv')
df
```

```python
# null values checking
df.isnull().sum()
```

```python
test1_mean = df['test1'].mean()
test2_mean = df['test2'].mean()

print("test1 mean is:",test1_mean)
print("test2 mean is:",test2_mean)
```

```python
df['test1'] = df['test1'].fillna(df['test1'].mean())
df['test2'] = df['test2'].fillna(df['test1'].mean())
```

```python
df.isnull().sum()
```

```python
# TASK 3
df.tuitionfee_waiver.value_counts()
```

```python
# replacing the tutionfee_waivers column with yes and no
df.replace({'Does not qualify': 'No'}, inplace=True)
df.replace({'Qualifies for reduced/free ': 'Yes'}, inplace=True)
```

```python
# TASK 4
g1 = df.groupby("school")['test1'].mean()  
# Calculate the mean test score for each school
overall_mean = df['test1'].mean()  
# Calculate the overall mean test score across all schools
name = df.groupby('school')['test1'].mean() == overall_mean 
# Determine which schools have the same mean test score as the overall mean

print('\n',g1)
#showing result
print('\nSchool with highest average test score:', g1.idxmax(), g1.max())
```

```python
# TASK 5
pvt = pd.pivot_table(df, index=['school'], values=['test2'],aggfunc='max')# Create a pivot table to find the maximum 'test2' score for each school
top_five_schools = pvt.sort_values(by = 'test2', ascending=False).head(5)   # Sort the pivot table to find the top five schools based on 'test2' scores
print("Top five schools based on 'test2' scores:")  # Print the top five schools
print(top_five_schools)
```

```python
# TASK 6
scholarship = pd.DataFrame({'scholarship': ['NA']*2133}, index = np.arange(0,2133))
scholarship
```
