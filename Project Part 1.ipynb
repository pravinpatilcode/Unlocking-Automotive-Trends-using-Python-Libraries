```python
# Define the list of driving modes to check
required_modes = ['Normal', 'Comfort', 'Eco', 'Sport', 'Power Mode']

# Function to check if all required modes are present in the 'Drive_Modes' column
def has_all_modes(modes_string, required_modes):
    if pd.isna(modes_string):  # Check for NaN values
        return False
    modes = modes_string.split(',')  # Assuming modes are comma-separated
    modes = [mode.strip() for mode in modes]  # Clean up spaces
    return all(mode in modes for mode in required_modes)

# Apply the function to filter the dataframe
filtered_df = df[df['Drive_Modes'].apply(lambda x: has_all_modes(x, required_modes))]

# List all details of cars that have all the required driving modes
print("Details of cars with all required driving modes:")
print(filtered_df)

# Count how many cars have all the required driving modes
num_cars_with_all_modes = filtered_df.shape[0]
print(f"\nNumber of cars with all the required driving modes: {num_cars_with_all_modes}")
```

```python
#------------------------------------ TASK 11

# Function to handle range values and convert kilograms to pounds
def preprocess_and_convert(column):
    def process_value(value):
        if isinstance(value, str):
            if 'Not Available' in value:
                return np.nan
            if '-' in value:
                # Handle range values by taking the average
                try:
                    min_val, max_val = value.split('-')
                    average_val = (float(min_val) + float(max_val)) / 2
                    return average_val
                except ValueError:
                    return np.nan
            if 'kg' in value:
                # Remove 'kg' and convert to float
                value = value.replace(' kg', '')
                try:
                    return float(value)
                except ValueError:
                    return np.nan
        return np.nan

    # Apply processing to each element
    column = column.apply(process_value)
    # Convert kilograms to pounds
    column = column * 2.2
    return column

# Apply preprocessing to both columns
df2['Kerb_weight'] = preprocess_and_convert(df2['Kerb_weight'])
df2['Gross_Vehicle_Weight'] = preprocess_and_convert(df2['Gross_Vehicle_Weight'])

print(df2)
```

```python
#------------------------------------ TASK 12

# Calculate statistics
def compute_statistics(group):
    mean = group.mean()
    median = group.median()
    std_dev = group.std()
    coeff_of_var = std_dev / mean if mean != 0 else np.nan
    return pd.Series({
        'Mean': mean,
        'Median': median,
        'Std Dev': std_dev,
        'Coeff of Variation': coeff_of_var
    })

# Apply the statistics calculation for each relevant column and group by manufacturer
statistics = {}
for column in ['Displacement', 'Cylinders', 'Valves_Per_Cylinder']:
    stats = df.groupby('Make')[column].apply(compute_statistics).unstack()
    statistics[column] = stats

# Combine all statistics into a single DataFrame
stats_df = pd.concat(statistics, axis=1)
print(stats_df)
```
